```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('DatasetDocBot_final.csv')
data.head(15)
```

```python
len(data.columns)
```

```python
# ei khane data ke 2 way te devide korbo
# input multiple hbe but output sudu ekta tai hbe
X = data.iloc[:,1:20].values
y = data.iloc[:,20].values
```

```python
X[250]
```

```python
y
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear', random_state = 0)
model.fit(X,y) 
```

```python
from sklearn.model_selection import KFold
n_splits=10
kf = KFold(n_splits=n_splits,random_state=0, shuffle=True)
```

```python
from sklearn.model_selection import cross_val_score
results = cross_val_score(model,X,y,cv=kf)
results
```

```python
results.mean() # Accuracy
```

```python
print("Accuracy: %.3f%% (%.3f%%)"% (results.mean()*100.0, results.std()*100.0))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

for train_index, val_index in kf.split(X):
    model.fit(X[train_index], y[train_index])
    pred = model.predict(X[val_index])
    print(confusion_matrix(y[val_index], pred))
    print(classification_report(y[val_index], pred))
```

```python
def predictDisease(array19):
    return model.predict([array19])[0]
```
